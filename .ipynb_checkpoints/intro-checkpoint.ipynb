# NLP Preprocessing Demo Notebook

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.util import ngrams
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, FreqDist
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Download necessary NLTK resources
nltk.download("punkt_tab", quiet=True)
nltk.download("stopwords", quiet=True)
nltk.download("wordnet", quiet=True)
nltk.download("averaged_perceptron_tagger_eng", quiet=True)
```

### Load dataset

```python
# Sample dataset: abstracts and titles
data = {
    "title": [
        "Deep Learning for Text Classification",
        "Natural Language Processing with Transformers",
        "An Overview of Neural Language Models",
    ],
    "abstract": [
        "Deep learning methods have proven very effective in text classification tasks. We explore various models and techniques.",
        "This paper explores the use of transformer-based models in various NLP tasks including classification and summarization.",
        "We review several neural network architectures for language modeling and their applications in NLP.",
    ],
}

df = pd.DataFrame(data)
df.head()
```

```python
# Helper function to clean and tokenize text
def preprocess_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)  # keep only letters and digits
    tokens = word_tokenize(text)  # split into words
    filtered_tokens = [word for word in tokens if word not in stop_words]  # drop stopwords
    return filtered_tokens
```

### Stopwords and stemmer/lemmatizer

**Stop words** are a set of commonly used words in any language. For example, in English, “the”, “is” and “and”, would easily qualify as stop words. In NLP and text mining applications, stop words are used to eliminate unimportant words, allowing applications to focus on the important words instead.

**Stemming** is the process of reducing a word to its root form by chopping off prefixes or suffixes, often without regard for grammar or meaning.
For example, "running", "runs", and "runner" might all be reduced to "run" or "runn".

It's a fast, rule-based method, but it can produce non-words and over-stem (e.g., "studies" → "studi"). Stemming is useful when speed matters more than precision.

See also https://vijinimallawaarachchi.com/2017/05/09/porter-stemming-algorithm/ .

**Lemmatization** is the process of reducing a word to its base or dictionary form (called a lemma), while considering the word’s part of speech and context.
For example, "running" becomes "run", and "better" becomes "good".

Unlike stemming, lemmatization produces real words and is more accurate, though typically slower. It's useful when you want to normalize words while preserving their meaning.

```python
# Define stopwords and stemmer/lemmatizer
stop_words = set(stopwords.words("english"))
print(list(stop_words)[:20])  # peek at some words
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

### 1. Average length of abstract (in words)

```python
df["abstract_tokens"] = df["abstract"].apply(word_tokenize)  # tokenize each abstract
print(df["abstract_tokens"])
df["abstract_len"] = df["abstract_tokens"].apply(len)
print("Average abstract length:", df["abstract_len"].mean())
```

### 2. Average sentence length in words (rough proxy for sentence complexity)


```python
def avg_sentence_length(text):
    sentences = sent_tokenize(text)  # sentence tokenization
    word_counts = [len(word_tokenize(sent)) for sent in sentences]
    return np.mean(word_counts)


df["avg_sentence_length"] = df["abstract"].apply(avg_sentence_length)
print("Average sentence length:", df["avg_sentence_length"].mean())
```

### 3. Frequency of non-stopwords

```python
all_tokens = df["abstract"].apply(preprocess_text).sum()
token_freq = Counter(all_tokens)
print("Most common non-stop words:", token_freq.most_common(10))
```

### 4. Common bi-grams and tri-grams (frequent word combinations)

Bigrams and trigrams are types of n-grams, which are sequences of n consecutive words from a text.

    A bigram is a pair of consecutive words (n=2), e.g., "new york", "machine learning".

    A trigram is a sequence of three words (n=3), e.g., "natural language processing".

They capture local word patterns and are useful in tasks like language modeling, text classification, and information retrieval, especially where word order matters.

```python
def get_ngrams(tokens, n):
    return list(ngrams(tokens, n))


all_bigrams = df["abstract"].apply(lambda x: get_ngrams(preprocess_text(x), 2)).sum()
all_trigrams = df["abstract"].apply(lambda x: get_ngrams(preprocess_text(x), 3)).sum()

bigram_freq = Counter(all_bigrams)
trigram_freq = Counter(all_trigrams)

print("Most common bigrams:", bigram_freq.most_common(5))
print("Most common trigrams:", trigram_freq.most_common(5))
```

### 5. Title structure analysis (common n-grams in titles)

```python
title_tokens = df["title"].apply(preprocess_text)
print(title_tokens, "\n")
all_title_ngrams = title_tokens.apply(lambda x: get_ngrams(x, 2)).sum()
title_bigram_freq = Counter(all_title_ngrams)
print("Common title bigrams:", title_bigram_freq.most_common(5))
```

### 6. Stemming and Lemmatization comparison (simplifying word forms)


```python
sample_text = "running runs ran runner"
tokens = word_tokenize(sample_text)
stemmed = [ps.stem(w) for w in tokens]
# POS must be specified (e.g., verb). You can also try incorrect values
lemmatized = [lemmatizer.lemmatize(w, pos="v") for w in tokens]  
print("Original:", tokens)
print("Stemmed:", stemmed)
print("Lemmatized:", lemmatized)
```

### 7. POS tagging (identifying the grammatical role of each word)

Part-of-speech tagging (POS tagging or PoS tagging or POST), also called grammatical tagging is the process of marking up a word in a text (corpus) as corresponding to a particular part of speech,[1] based on both its definition and its context. In the English language, there are 9 parts of speech commonly listed: noun, verb, article, adjective, preposition, pronoun, adverb, conjunction, and interjection. However, there are many more categories and sub-categories. For example, NN for singular common nouns, NNS for plural common nouns, NP for singular proper nouns (see the POS tags used in the Brown Corpus).

For more details, see https://en.wikipedia.org/wiki/Part-of-speech_tagging .

```python
sample_abstract = df["abstract"][0]
tokens = word_tokenize(sample_abstract)
pos_tags = pos_tag(tokens)
print("POS Tags (first abstract):", pos_tags[:10])

# POS frequency (e.g., how many nouns, verbs, etc.)
all_pos = df["abstract"].apply(lambda x: pos_tag(word_tokenize(x))).sum()
pos_freq = FreqDist(tag for word, tag in all_pos)
pos_freq.plot(10, title="Top POS tags")
```

### 8. Vocabulary size and frequency distribution


```python
fdist = FreqDist(all_tokens)
print("Vocabulary size:", len(fdist))
print(fdist)
fdist.plot(20, title="Top 20 words in corpus")
```

### 9. Converting text into feature vectors (Bag-of-Words and TF-IDF)

The "bag of words" model is a method in NLP and information retrieval that treats text as an unordered collection of words, focusing on their frequency and ignoring word order and context.

```python
# Bag-of-Words (count each word occurrence)
count_vectorizer = CountVectorizer(stop_words="english")  # Convert a collection of text documents to a matrix of token counts.
bow_matrix = count_vectorizer.fit_transform(df["abstract"])  # has shape (n_documents X n_features)
print("BoW feature matrix shape:", bow_matrix.shape)
print("Feature names (vocabulary):", count_vectorizer.get_feature_names_out()[:10])
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())
```

**TF-IDF (Term Frequency–Inverse Document Frequency)** is a numerical statistic that measures the importance of a word in a document relative to a collection of documents (corpus). It combines two ideas:

- **Term Frequency (TF)**: How often a word appears in a single document.
- **Inverse Document Frequency (IDF)**: How unique or rare that word is across the entire corpus.

This helps to reduce the weight of common words (like "the", "is") and emphasize rarer, more informative terms.

The formula for TF-IDF of a term \\( t \\) in a document \\( d \\) is:

\\[
\text{TF-IDF}(t, d) = \text{TF}(t, d) \times \log \left( \frac{N}{1 + \text{DF}(t)} \right)
\\]

Where:
- \\( \text{TF}(t, d) \\): frequency of term \\( t \\) in document \\( d \\)
- \\( N \\): total number of documents
- \\( \text{DF}(t) \\): number of documents containing term \\( t \\)

> Note: The "+1" in the denominator avoids division by zero.


```python
# TF-IDF (weigh common words lower, highlight rare informative ones)
tfidf_vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf_vectorizer.fit_transform(df["abstract"])
print("\nTF-IDF feature matrix shape:", tfidf_matrix.shape)
print("TF-IDF feature names:", tfidf_vectorizer.get_feature_names_out()[:10])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

Visualize results

```python
# Display side-by-side
print("Bag-of-Words Matrix:")
display(bow_df)

print("TF-IDF Matrix:")
display(tfidf_df)
```
